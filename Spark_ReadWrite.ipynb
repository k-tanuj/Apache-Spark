```python
path = "/data/messy_customer_data.csv"
```

# SPARK READ

### 1. FAIL FAST - STOPS WHOLE EXECUTION IF SOME INVALID VALUE COMES

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("CustomerID", StringType(), True),
    StructField("Name", StringType(), True),
    StructField("Gender", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("City", StringType(), True),
    StructField("Signup_Date", StringType(), True),
    StructField("Last_purchase_date", StringType(), True),
    StructField("purchase_amount", DoubleType(), True),
    StructField("feedback_score", IntegerType(), True),
    StructField("email", StringType(), True),
    StructField("Phone_number", StringType(), True),
    StructField("Country", StringType(), True)
])
```

```python
df_failFast = spark.read\
.format('csv')\
.option('header','true')\
.option('mode','failfast')\
.schema(schema)\
.load(path)
df_failFast.show()
```

### 2.PERMISSIVE - malformed records are processed, allow us to inspect records without stopping execution

```python
df_permissive = spark.read\
.format('csv')\
.option('header','true')\
.option('mode','PERMISSIVE')\
.schema(schema)\
.load(path)
df_permissive.show()
```

### 3. DROP MALFORMED - automatically removes the malformed records. no error, no warning, and no trace of traced records.

```python
df_mal = spark.read\
.format('csv')\
.option('header','true')\
.option('mode','DROPMALFORMED')\
.schema(schema)\
.load(path)
df_mal.show()
```

# SPARK WRITE

```python
data = [
    (1, "Ravi", 25, "Mumbai", 45000),
    (2, "Ankit", 30, "Delhi", 55000),
    (3, "Priya", 22, "Bangalore", 48000),
    (4, "Neha", 28, "Jaipur", 52000),
    (5, "Aman", 35, "Pune", 60000)
]

columns = ["id", "name", "age", "city", "salary"]
```

```python
df = spark.createDataFrame(data,columns)
```

```python
df.show()
```

```python
df.repartition(1).write\
.format('csv')\
.option('delimiter','|')\
.option('header','true')\
.save('/data/write_output1.csv')
```
