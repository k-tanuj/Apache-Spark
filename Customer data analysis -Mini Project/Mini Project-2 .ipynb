```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master("yarn").appName("CustomerDataProcessing").getOrCreate()
spark
```

```python
orders_df = spark.read.format('csv').option('header','true').load('/data/orders.csv')

orders_df.show(5)
```

```python
orders_df.printSchema()
```

```python
from pyspark.sql.functions import col

orders_df = orders_df.withColumn("order_id", col("order_id").cast("int")) \
       .withColumn("customer_id", col("customer_id").cast("int")) \
       .withColumn("order_date", col("order_date").cast("date")) \
       .withColumn("total_amount", col("total_amount").cast("double"))

orders_df.printSchema()
```
